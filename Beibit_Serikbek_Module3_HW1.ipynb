```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
import urllib
urllib.request.urlretrieve('https://github.com/zbz95/Intro_to_ML/raw/main/insurance.csv', 'insurance.csv')
```

# Упражнение 1
Откройте файл insurance.csv

Поделите файл на три блока train, validation, test. Размер train-80% всех данных, validation 10% всех данных, test 10% всех данных

Стандартизируйте features во всех трех блоках используя StandardScaler

```python
df = pd.read_csv('insurance.csv')

display(df)

train,validation = train_test_split(df, test_size = 0.2, shuffle = True)

test,validation = train_test_split(validation, test_size = 0.5, shuffle = True)

scaler = StandardScaler()

scaler.fit(train[['age', 'bmi', 'children', 'sex', 'smoker']])
scaler.fit(validation[['age', 'bmi', 'children','sex', 'smoker']])
scaler.fit(test[['age', 'bmi', 'children', 'sex', 'smoker']])

train[['age', 'bmi', 'children', 'sex', 'smoker']] = scaler.transform(train[['age', 'bmi', 'children', 'sex', 'smoker']])
validation[['age', 'bmi', 'children', 'sex', 'smoker']] = scaler.transform(validation[['age', 'bmi', 'children', 'sex', 'smoker']])
test[['age', 'bmi', 'children', 'sex', 'smoker']] = scaler.transform(test[['age', 'bmi', 'children', 'sex', 'smoker']])

display(train)
display(validation)
display(test)
```

# Упражнение 2

Постройте модель линейной регрессии(LinearRegression) на блоке train и посчитайте root_mean_squared_error на блоке validation

```python
model_1 = LinearRegression()
model_1.fit(train[['age', 'bmi', 'children', 'smoker', 'sex']], train[['expenses']])

model_1.coef_, model_1.intercept_

validation['prediction_1'] = model_1.predict(validation[['age', 'bmi', 'children', 'smoker', 'sex']])

root_mean_squared_error(validation['expenses'], validation['prediction_1'])
```

# Упражнение 3
Постройте модели KNeighborsRegressor используя разные варианты числа соседей, k = [5,10,15,20,25,50]

Обучите все модели на train и посчитайте  root_mean_squared_error на блоке validation для каждого k

```python
for k in [5,10,15,20,25,50]:
  model_2 = KNeighborsRegressor(n_neighbors = k)
  model_2.fit(train[['age', 'bmi', 'children', 'smoker', 'sex']], train['expenses'])

  validation[f'prediction_{k}'] = model_2.predict(validation[['age', 'bmi', 'children', 'smoker', 'sex']])
  rmse1 = root_mean_squared_error(validation['expenses'], validation[f'prediction_{k}'])

  print(f'Number of nearest neighbors = {k}. Root mean squared error = {rmse1}')
```

# Упражнение 4

Постройте модели KNeighborsRegressor используя разные наборы features. Например используя сначала все features, после исключая feature одну за другой.

Обучите все модели на train и посчитайте root_mean_squared_error на блоке validation для каждого набора

```python
features_list = [['age', 'bmi', 'children', 'smoker', 'sex'],
                 ['age', 'bmi', 'children', 'smoker'],
                 ['age', 'bmi', 'children'],
                 ['age', 'bmi'],
                 ['age']]

for features in features_list:
  model_3 = KNeighborsRegressor(n_neighbors = 10)
  model_3.fit(train[features], train['expenses'])

  validation['prediction'] = model_3.predict(validation[features])
  rmse2 = root_mean_squared_error(validation['expenses'], validation['prediction'])

  print(f'Used features = {features}. Root mean squared error = {rmse2}')
```

# Упражнение 5

Определите модель у которой был самый низкий  root_mean_squared_error на блоке validation. Проверьте root_mean_squared_error данной модели на блоке test

```python
model_4 = KNeighborsRegressor(n_neighbors = 10)
model_4.fit(train[['age', 'bmi', 'children', 'smoker']], train['expenses'])

validation['prediction'] = model_4.predict(validation[['age', 'bmi', 'children', 'smoker']])
rmse3_1 = root_mean_squared_error(validation['expenses'], validation['prediction'])

print(f' Root mean squared error on validation data = {rmse3_1}')


test['prediction'] = model_4.predict(test[['age', 'bmi', 'children', 'smoker']])
rmse3_2 = root_mean_squared_error(test['expenses'], test['prediction'])

print(f' Root mean squared error on test data = {rmse3_2}')
```
