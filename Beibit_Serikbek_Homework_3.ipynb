9. Построите график на среднюю заработную плату ('salary_in_usd') по типу занятости и уровню опыта ('experience_level')

```python
data2 = data.groupby(['experience_level'], as_index=False).agg(

  avg_salary = ('salary_in_usd', 'mean')

)

data2.sort_values('avg_salary', inplace = True)

plt.bar(data2['experience_level'], data2['avg_salary'])

plt.title("Average Salary Across Different Experience Levels")

plt.ylabel('Average Salary (USD)')
plt.xlabel('Experience Level')

```
