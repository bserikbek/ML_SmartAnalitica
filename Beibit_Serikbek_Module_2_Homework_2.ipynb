Если 'language' содержит 'Python' , то 'web development' должен содержать слово 'Full stack'.

Если 'language' содержит 'PHP' , то 'web development' должен содержать слово 'Full stack'.

Если 'language' содержит 'Java' , то 'web development' должен содержать слово 'Backend'.

Если 'language' содержит 'CSS' , то 'web development' должен содержать слово 'Frontend'.

```python
import numpy as np

def devType(sString):
  if sString in ['JavaScript', 'HTML', 'CSS']:
    return 'Frontend'
  elif sString in ['Java', 'Go']:
    return 'Backend'
  elif sString in ['Python', 'PHP']:
    return 'Full stack'
  else: return np.nan


df['web development'] = df['language'].map(devType)

print(df.shape)

df.loc[df['web development'].isin(['Full stack', 'Frontend', 'Backend'])]


```

15. В приведенном ниже словаре ключами являются height, а значениями — кортежи нижнего и верхнего предела идеального веса соответственно.

Если кто-то ниже идеального веса для своего роста, он «Thin»(то есть Худой). Если кто-то выше идеального веса для своего роста, он «Obese» (то есть Ожиревший). Если кто-то находится в диапазоне идеального веса для своего роста, он  «Fit» (то есть Подтянутый).

Смотря на эти правила напишите функцию, которая заполнит по этим критериям новый столбец "Body Type" в нижеуказанном дата фрейме

```python
import pandas as pd
df = pd.DataFrame(
         {
             'Name': ['Edward', 'Natalie', 'Chris M', 'Priyatham'],
             'Sex' : ['M', 'F', 'M', 'M'],
             'Age': [45, 35, 29, 26],
             'weight(kgs)': [68.4, 58.2, 64.3, 53.1],
             'height(cms)' : [178, 160, 173, 168]
         }
     )
df
```

```python
def bodyType(height, weight, sex):

  male_fitness = {
     #height  : (weight_lower_cap, weight_upper_cap)
     178      : (     67.5       ,      83         ),
     173      : (     63         ,      70.6       ),
     168      : (     58         ,      70.7       )
 }

  female_fitness = {
     #height  : (weight_lower_cap, weight_upper_cap)
     160      : (     47.2       ,      57.6       )
 }

  if sex == 'M':
    if male_fitness[height][0] > weight:
      return 'Thin'
    elif male_fitness[height][1] < weight:
      return 'Obese'
    else: return 'Fit'
  else:
    if female_fitness[height][0] > weight:
      return 'Thin'
    elif female_fitness[height][1] < weight:
      return 'Obese'
    else: return 'Fit'
```

```python
df['Body Type'] = df.apply(lambda row: bodyType(row['height(cms)'], row['weight(kgs)'], row['Sex']), axis = 1)

df
```
